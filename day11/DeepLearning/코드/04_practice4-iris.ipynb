```python
# 라이브러리 사용
import pandas as pd
import tensorflow as tf
```

```python
# 1.과거의 데이터를 준비합니다.
파일경로 = './csv/iris.csv'
아이리스 = pd.read_csv(파일경로)
아이리스.head()
```

```python
# 원핫인코딩
인코딩 = pd.get_dummies(아이리스)
인코딩.head()
```

```python
# 칼럼이름 출력
print(인코딩.columns)
```

```python
# 독립변수, 종속변수
독립 = 인코딩[['꽃잎길이', '꽃잎폭', '꽃받침길이', '꽃받침폭']]
종속 = 인코딩[['품종_setosa', '품종_versicolor', '품종_virginica']]
print(독립.shape, 종속.shape)
```

```python
# 2. 모델의 구조를 만듭니다
X = tf.keras.layers.Input(shape=[4]) # 독립변수 13개 이므로 13
Y = tf.keras.layers.Dense(3, activation='softmax')(X)  # activation='softmax' : 분류 모델중 softmax 사용하겠다.
model = tf.keras.models.Model(X, Y)
# 다중분류손실함수 : categorical_crossentropy는 데이터의 라벨이 원핫인코딩 된 형태의 경우 사용. 
# 보통 softmax 활성화함수 다음에 연계되어 사용되므로 softmax loss로도 불린다.
# sparse_categorical_crossentropy는 데이터의 라벨이 Int형 일 때 사용한다.
model.compile(loss='categorical_crossentropy',  # loss: 손실 함수를 설정합니다. 모델이 예측한 값과 실제 값 사이의 차이를 계산하는 함수입니다.
              metrics = ['accuracy'])
# metrics: 평가 지표를 설정합니다.학습 중 모델의 성능을 평가할 때 사용됩니다. 
# 정확도(Accuracy), 정밀도(Precision), 재현율(Recall), F1 점수 등이 있습니다.
```

```python
# 3.데이터로 모델을 학습(FIT)합니다.
model.fit(독립, 종속, epochs=100)
```

```python
# 모델을 이용합니다.
# 맨 처음 데이터 5개
print(model.predict(독립[:5]))
print(종속[:5])  # 실제 정답
```

```python
# 맨 끝 데이터 5개
print(model.predict(독립[-5:]))
print(종속[-5:])  # 실제 정답
```

```python
# weights & bias 출력. 학습한 가중치
print(model.get_weights())
```

첫번째 컬럼은 바로 setosa를 예측하는 수식이라고 보시면 됩니다. 첫번째 -0.35968933은 꽃잎의 길이와 곱해주는 가중치이고, 두번째1.2256223은 꽃잎의 폭에 곱해주는 가중치, 그리고 세번째  -0.3401752 는 꽃받침의 길이를 곱해주는 가중치, 네번째 인 -1.589324 은 꽃받침의 폭을 곱해주는 가중치입니다. 그래서 이렇게 수식을 하나 만들고 마지막에 bias 인 0.8977075을 더해주면 setosa인지 아닌지를 가늠할 수 있는 값이 나오는 것입니다. 여기에 softmax를 한번 씌워주면 그 확률이 나와서 아까 보셨듯이 0.8***뭐 이렇게 나왔던 거죠. 그리고 두번째 것은 versicolor인째 컬럼이 virginica일 확률에 대해서 계산해주는 수식에 이렇게 가중치를 나타내는 것입니다. 여러분들이 이렇게 종속변수가 범주형일때, 분류 모델을 만드는 법을 학습을 했습니다. 그리고 이런 범주형 데이터를 학습시키기 위해서는 수치형으로 변형이 필요한데 그변형시켜 주는 방법까지 학습이 끝났습니다.
