```python
# from langchain.chains import RetrievalQA
# from langchain.chat_models import ChatOpenAI
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.vectorstores import Chroma
from langchain.document_loaders import PyPDFLoader
```

```python
loader = PyPDFLoader("/content/drive/MyDrive/파이썬/ChatBot/[이슈리포트 2022-2호] 혁신성장 정책금융 동향.pdf")
pages = loader.load_and_split() # 로드엔 스플릿으로 페이지별로 나눈다후 결과값 저장.
```

# 새 섹션

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
# 리시버캐릭터 텍스트 스필리터로 텍스트를 분할 청크사이즈(얼마나 작게 자를 건지)는 500,  chunk_overlap : 잘랐을 때 얼마나 겹치게 가져갈 것인지.
texts = text_splitter.split_documents(pages) # 저장

from langchain.embeddings import HuggingFaceEmbeddings

model_name = "jhgan/ko-sbert-nli" # 한글이므로 이모델 선택
model_kwargs = {'device': 'cpu'}
encode_kwargs = {'normalize_embeddings': True}
hf = HuggingFaceEmbeddings(
    model_name=model_name,
    model_kwargs=model_kwargs,
    encode_kwargs=encode_kwargs
)

docsearch = Chroma.from_documents(texts, hf) # 크로마DB에 저장
```

```python
 retriever = docsearch.as_retriever( # 검색기능 활용
                                    search_type="mmr", # 최대한 다양한 답변을 할수 있게 선언
                                    search_kwargs={'k':3, 'fetch_k': 10}) # 총 10개 유사 문서를 가장 다양하게 조합하도록 3개만 뽑아라
 retriever.get_relevant_documents("혁신성장 정책금융에 대해서 설명해줘")
```

Answer the question as based only on the following context 주어진 맥락에서 답변을 해라.

```python
from langchain.prompts import ChatPromptTemplate # 쳇 프롬프트 템플릿을 통해 우회적으로 레그시스템을 구성하는 것을 함.
from langchain.schema.runnable import RunnableMap

template = """Answer the question as based only on the following context:
{context}

Question: {question}
"""

prompt = ChatPromptTemplate.from_template(template)
```

```python
gemini = ChatGoogleGenerativeAI(model="gemini-pro", temperature = 0) # 템플레이트 0으로 설정해야 답변의 일과성을 줄수 있음.

chain = RunnableMap({
    "context": lambda x: retriever.get_relevant_documents(x['question']),
    "question": lambda x: x['question']
}) | prompt | gemini
```

```python
Markdown(chain.invoke({'question': "혁신성장 정책금융에 대해서 설명해줘"}).content)
```
